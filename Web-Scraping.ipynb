```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
ur1='https://www.imdb.com/search/title/?groups=top_250&sort=user_rating,desc&start=51&ref_=adv_nxt'
page=requests.get(ur1)
soup=BeautifulSoup(page.text,'html.parser')
```

```python
id=[]
id_elem=soup.find_all(class_='lister-item-index unbold text-primary')
for item in id_elem:
    id.append(item.text)     
```

```python
year=[]
year_elem=soup.find_all(class_='lister-item-year text-muted unbold')
for item in year_elem:
    year.append(item.text)    
```

```python
type=[]
type_elem=soup.find_all(class_='genre')
for item in type_elem:
    type.append(item.text)    
```

```python
meta_score=[]
meta_score_elem=soup.find_all(class_='inline-block ratings-metascore')
for item in meta_score_elem:
    meta_score.append(item.text)      
```

```python
summary_text=[]
summary_text_elem=soup.find_all(class_='text-muted')
for item in summary_text_elem:
    summary_text.append(item.text)  
    
```

```python
votes=[]
votes_elem=soup.find_all(class_='sort-num_votes-visible')
for item in votes_elem:
    votes.append(item.text)     
```

```python
final_array=[]
for id,year,gener,metascore,summary,votes_and_gross in zip(id,year,type,meta_score,summary_text,votes):
    final_array.append({'id':id,'year':year,'type':gener,'meta_score':metascore,'summary_text':summary,'votes':votes_and_gross})
df=pd.DataFrame(final_array)
df
```
